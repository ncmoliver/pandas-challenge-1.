## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df[['unit_price', 'qty', 'subtotal']]
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight'] = df['qty'] * df['unit_weight']
df['shipping_price'] = df['total_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)
df[['qty', 'unit_weight', 'total_weight', 'shipping_price']]
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
df['total_price'] = round((df['shipping_price'] + df['subtotal']) * 1.0925, 2)
df[['subtotal', 'shipping_price', 'total_price']]


```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df['cost_of_each_line'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df[['first', 'last', 'unit_cost', 'qty', 'shipping_price', 'cost_of_each_line']]
```

```python
# Create a column for the profit of each line using line cost and line price
df['profit_of_each_line'] = df['total_price']  - df['cost_of_each_line']
df[['first', 'last', 'unit_price', 'cost_of_each_line', 'profit_of_each_line']]
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
def quantity_spent(client_id, column):
    top_5_spending = df.loc[df['client_id'] == client_id, column] 
    return round(top_5_spending.sum(), 2)

for x in top_5_list:
    print(f"{x} the total is {quantity_spent(x, 'shipping_price')}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
def calculate_totals_for_top_clients(df, top_clients):
    results = []
    for client_id in top_clients:
        client_data = df[df['client_id'] == client_id]
        total_units_purchased = client_data['qty'].sum()
        total_shipping_price = client_data['shipping_price'].sum()
        total_revenue = client_data['total_price'].sum()
        total_profit = client_data['profit_of_each_line'].sum()
        results.append([client_id, total_units_purchased, total_shipping_price, total_revenue, total_profit])
    
    summary_df = pd.DataFrame(results, columns=['client_id', 'total units purchased', 'total shipping price', 'total revenue', 'total profit'])
    return summary_df

new_dataFrame = calculate_totals_for_top_clients(df, top_5_list)
new_dataFrame
```

```python
# Format the data and rename the columns to names suitable for presentation.
new_dataFrame = new_dataFrame.rename(columns={
    'client_id' : 'Client ID',
    'total units purchased' : 'Units Purchased'
}).copy()
# Define the money columns.
money_columns = new_dataFrame[['total shipping price', 'total revenue', 'total profit']]
# # Define a function that converts a dollar amount to millions.
def dollar_to_millions(x):
    return f"${x/1000000:.2f} M"

# Apply the currency_format_millions function to only the money columns. 
money_to_millions = money_columns.map(dollar_to_millions)
money_to_millions
# Rename the columns to reflect the change in the money format. 
renamed_df = money_to_millions.rename(columns={
    'total shipping price' : 'Shipping Price (in millions)',
    'total revenue':'Revenue (in millions)',
    'total profit':'Profit (in millions)',
})

renamed_df
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
sorted_df = renamed_df.sort_values("Profit (in millions)", ascending=False)
updated_df = sorted_df.reset_index(drop=True).copy()
columns_to_add = new_dataFrame[['Client ID', 'Units Purchased']]
final_df = pd.concat([columns_to_add, updated_df], axis=1).copy()
final_df
```
